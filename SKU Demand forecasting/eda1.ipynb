## Filtering essential columns

```python
df2.head(3)
```

```python
df3 = df2[['Date','InvoiceId','Qty','Freight','LineAmount1','InvoiceCat','CustomerName','State','City']]
```

```python
df3['Year'] = df3['Date'].apply(lambda x: str(x).split('-')[0])
df3['Month'] = df3['Date'].apply(lambda x: str(x).split('-')[1])
df3['Day'] = df3['Date'].apply(lambda x: str(x).split('-')[2][:2])
df3['Y-M'] = df3['Date'].apply(lambda x: '-'.join(str(x).split('-')[:2]))
df3['WeekDay'] = df3['Date'].apply(lambda x: datetime.datetime.date(x).strftime('%A'))
```

```python
df3.head(3)
```

```python
def get_describe_numeric(coln,df_x=df3,f_coff=0):
    for x in sorted(list(set(df_x[coln]))):
        df_y = df_x[df_x[coln]==x]
        if df_y.shape[0]>f_coff:
            print(x)
            print(df_y.describe())
            print()
            print(df_y.corr())
            print("-------------------------------------------")
```

```python
get_describe_numeric("InvoiceCat")
```
